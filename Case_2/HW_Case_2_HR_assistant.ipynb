Ваше задание сделать так, чтобы нейронная сеть могла делать свой прогноз на основе не 4, а 10 параметров. При этом можно провести собственный опрос для оценки весов, а можно выбрать параметры и весовые коэффициенты, представленные в файле **case2.csv**	

```python
#импорт библиотек
import pandas as pd
import numpy as np
```

```python
class NeuralNetwork():
    # Метод init инициализирует веса и смещение нейронной сети из файла weights_file, который по умолчанию равен 'case2.csv'
    def __init__(self, weights_file='case2.csv', bias=0):
        weights_df = pd.read_csv(weights_file, header=None, index_col=0)
        self.weights = np.append(weights_df[1].values, 1)
        self.bias = bias
        self.inputs = np.zeros(len(self.weights))
        self.hmin = 1 / (1 + np.exp(-self.bias))
        self.hmax = 1 / (1 + np.exp(1 - self.bias))
    # Метод set_inputs запрашивает у пользователя входные данные, для чего для каждого вопроса, содержащегося в файле weights_file, выводится запрос на ввод оценки по шкале от 0 до 1
    def set_inputs(self):
        print('Ответьте на вопросы по стажировке (оцените условия по шкале от 0 до 1)')
        weights_df = pd.read_csv('case2.csv', header=None, index_col=0)
        for i, question in enumerate(weights_df.index):
            x = float(input(question + '? '))
            self.inputs[i] = x
        self.inputs[-1] = self.bias
    # Метод forward производит прямой проход через нейронную сеть, используя входные данные и веса, и вычисляет выход скрытого слоя до и после активации
    def forward(self):
        self.z = self.inputs @ self.weights
        print('Выход скрытого слоя до активации:', round(self.z, 3))
        self.h = 1 / (1 + np.exp(-self.z))
        print('Выход скрытого слоя после активации:', round(self.h, 3))
    # Метод normalize вычисляет вероятность соответствия выходных данных исходной шкале от 0 до 1
    def normalize(self):
        self.p = (self.h - self.hmin) / (self.hmax - self.hmin)
        self.p = max(0, min(self.p, 1)) # ограничиваем вероятность между 0 и 1
        print('Вероятность соответствия:', round(self.p, 3))
```

```python
nn = NeuralNetwork() 
```

```python
nn.set_inputs() 
```

```python
nn.forward() 
```

```python
nn.normalize() 
```
