```python
#подключаем библиотеки
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
import nltk
```

```python
# Считываем данные из файла
with open("case3.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
# Удаляем символы новой строки и приводим к нижнему регистру
text = text.replace('\n', ' ').lower()
```

```python
# Удаляем пунктуационные символы
punct = ['.', ',', ':', ';', '!', '?', '—', '-', '(', ')']
for p in punct:
    text = text.replace(p, '')
```

```python
# Удаляем стоп-слова
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('russian'))
text = ' '.join([word for word in text.split() if word not in stop_words])
```

```python
# Создаем объект WordCloud
wordcloud = WordCloud(background_color='white', width=1200, height=1000, random_state=42, max_words=100, max_font_size=200)
```

```python
# Создаем изображение Облака слов
wordcloud.generate(text)
```

```python
# Визуализируем Облако слов
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
